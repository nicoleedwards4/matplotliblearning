```python
import pandas as pd
colnames=['DATE', 'TAG', 'POSTS']
df=pd.read_csv('QueryResults.csv', names=colnames, header=None)
print(df)
```

```python
##Which programming language had the most number of posts since the creation of Stack Overflow
sum_by_tag = df.groupby('TAG').sum()
sum_by_tag = sum_by_tag.sort_values('POSTS', ascending=False)
print(sum_by_tag)
```

```python
##how many months of entries exist per programming language
count_by_tag = df.groupby('TAG').count()
count_by_tag = count_by_tag.sort_values('POSTS', ascending=False)
print(count_by_tag)
```
